```python
#capturing from webcam - best.pt
#importing required libraries
import cv2
from PIL import Image
from ultralytics import YOLO

model = YOLO("best.pt")  #loading the trained model

#open webcam for capturing feed
video_capture = cv2.VideoCapture(1)

while True:
    #read the live feed frame by frame 
    ret, frame = video_capture.read()
    if not ret: #if no frame, then exit
        break

    #predict on the input frame
    results = model.predict(frame, save=True, conf=0.60)
    boxes = results[0].boxes.xyxy #saving bounding boxes coordinates if object detected
    names = model.names
        
    #to draw bounding boxes on the frame
    
    for r in results:
        for box, c in zip(r.boxes.xyxy, r.boxes.cls):
            x_min, y_min, x_max, y_max = box.tolist()
            cv2.rectangle(frame, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0, 255, 0), 2)

            #predicted class label on top of bounding box
            label_text = names[int(c)] 
            cv2.putText(frame, label_text, (int(x_min), int(y_min) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    #displaying the frame in real time with bouding boxes
    cv2.imshow('Video',frame)

    #Check for exit key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture
video_capture.release()
cv2.destroyAllWindows()
```
